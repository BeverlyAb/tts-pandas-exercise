### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_csv(url, sep='\t')
```

### Step 4. How many products cost more than $10.00?

```python
chipo.head(1)
chipo['dollar_price'] = [float(val[1:]) for val in chipo['item_price']]

# delete duplicates in item_name, quantity, and choice_description 
chipo_no_dup = chipo.drop_duplicates(['item_name','quantity','choice_description'])

# only get orders with single items
chipo_single_item = chipo_no_dup[chipo_no_dup['quantity'] == 1]

# items_over_10 = [item for item in filter(lambda x : x > 10, chipo_single_item['dollar_price'])]
items_over_10 = chipo_single_item[chipo_single_item['dollar_price'] > 10]
items_over_10 = items_over_10.drop_duplicates('item_name')
print(f"The number of items costing over $10 is {len(items_over_10)}")
items_over_10
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
item_df = chipo_single_item[['item_name','item_price']].drop_duplicates('item_name')
item_df
```

### Step 6. Sort by the name of the item

```python
item_df.sort_values('item_name')
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
# Is the question, "How many times was the most expensive item ordered?"
item_df['item_price'] = [float(val) for val in item_df['item_price']]
item_df = item_df.sort_values('item_price', ascending=False)
most_expensive_item = item_df.head(1)['item_name'].array[0]
print(f"The most expensive item is {most_expensive_item}.")
most_exp_df = chipo_no_dup[chipo_no_dup['item_name'] == most_expensive_item]
print(f"This was ordered {most_exp_df['quantity'].sum()} times.")
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
# this will probably be better with regex
veg_df = chipo[chipo['item_name'] == 'Veggie Salad Bowl']
print(f"This was ordered {veg_df['quantity'].sum()} times.")
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
soda_df = chipo[(chipo['item_name'] == 'Canned Soda') & (chipo['quantity'] > 1)]
len(soda_df)
```
